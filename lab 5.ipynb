```python
import pandas as pd
from sklearn.linear_model import Perceptron
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
```

## 1. Загрузите обучающую и тестовую выборки из файлов perceptron-train.csv и perceptron-test.csv. Целевая переменная записана в первом столбце, признаки - во втором и третьем

```python
data_train = pd.read_csv('perceptron-train.csv')
data_test = pd.read_csv('perceptron-test.csv')
```

```python
X_train = data_train.iloc[:, 1:]
y_train = data_train.iloc[:, 0]

X_test = data_test.iloc[:, 1:]    
y_test = data_test.iloc[:, 0] 
```

```python
clf = Perceptron(random_state=241)
clf.fit(X_train, y_train)

X = clf.predict(X_test)

accuracy = accuracy_score(y_test, X)
accuracy
```

```python
scaler = StandardScaler()
X_train_new = scaler.fit_transform(X_train)
X_test_new = scaler.fit_transform(X_test)
```

```python
clf = Perceptron(random_state=241)
clf.fit(X_train_new, y_train)
X_scal = clf.predict(X_test_new)
accuracy_new= accuracy_score(y_test, X_scal)
accuracy_new
```
